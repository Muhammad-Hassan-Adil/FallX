```python
# FOR SINGLE FILE (TESTING)


import pandas as pd
import numpy as np
import os

#load csv to dataframe
df = pd.read_csv(r"C:\Users\Leonm\Desktop\FYP\Data\UMAFall_Subject_01_ADL_Aplausing_1_2017-04-14_23-38-231.csv", header=None)

# Drop the first 41 rows
df = df.iloc[41:]

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

# Split the data by semicolon into separate columns
df = df[0].str.split(';', expand=True)

# Drop the first column as it is not needed
df.drop(columns=[0], inplace=True)

#rename columns
df.columns = ['Time Stamp', 'X-Axis', 'Y-Axis', 'Z-Axis', 'Sensor Type', 'Sensor ID']

#drop row if Sensor ID is not 1 or 2
df['Sensor ID'] = pd.to_numeric(df['Sensor ID'], errors='coerce')
df = df[df['Sensor ID'].isin([1, 2])]
df['Time Stamp'] = pd.to_numeric(df['Time Stamp'], errors='coerce')
df['X-Axis'] = pd.to_numeric(df['X-Axis'], errors='coerce')
df['Y-Axis'] = pd.to_numeric(df['Y-Axis'], errors='coerce')
df['Z-Axis'] = pd.to_numeric(df['Z-Axis'], errors='coerce')
df['Sensor Type'] = pd.to_numeric(df['Sensor Type'], errors='coerce')

#split to new df by Sensor Type
df0 = df[df['Sensor Type'] == 0]
df1 = df[df['Sensor Type'] == 1]
df2 = df[df['Sensor Type'] == 2]

df0.reset_index(drop=True, inplace=True)
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

#split to new df by Sensor ID
df0_1 = df0[df0['Sensor ID'] == 1]
df0_2 = df0[df0['Sensor ID'] == 2]
df1_1 = df1[df1['Sensor ID'] == 1]
df1_2 = df1[df1['Sensor ID'] == 2]
df2_1 = df2[df2['Sensor ID'] == 1]
df2_2 = df2[df2['Sensor ID'] == 2]

#reset index
df0_1.reset_index(drop=True, inplace=True)
df0_2.reset_index(drop=True, inplace=True)
df1_1.reset_index(drop=True, inplace=True)
df1_2.reset_index(drop=True, inplace=True)
df2_1.reset_index(drop=True, inplace=True)
df2_2.reset_index(drop=True, inplace=True)

#rename df0_1 columns
df0_1.columns = ['Time Stamp','A1 X-Axis', 'A1 Y-Axis', 'A1 Z-Axis', 'Sensor Type', 'Sensor ID']
df0_2.columns = ['Time Stamp','A2 X-Axis', 'A2 Y-Axis', 'A2 Z-Axis', 'Sensor Type', 'Sensor ID']
df1_1.columns = ['Time Stamp','G1 X-Axis', 'G1 Y-Axis', 'G1 Z-Axis', 'Sensor Type', 'Sensor ID']
df1_2.columns = ['Time Stamp','G2 X-Axis', 'G2 Y-Axis', 'G2 Z-Axis', 'Sensor Type', 'Sensor ID']
df2_1.columns = ['Time Stamp','M1 X-Axis', 'M1 Y-Axis', 'M1 Z-Axis', 'Sensor Type', 'Sensor ID']
df2_2.columns = ['Time Stamp','M2 X-Axis', 'M2 Y-Axis', 'M2 Z-Axis', 'Sensor Type', 'Sensor ID']


#drop Sensor Type and Sensor ID columns
df0_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
df0_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
df1_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
df1_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
df2_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
df2_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)



#concatenate all dataframes
result = pd.concat([df0_1, df1_1, df2_1, df0_2, df1_2, df2_2], axis=1)
result
#create Time Stamp column from 1 and make it first column
result.insert(0, 'Time Stamp', range(1, 1 + len(result)))
result = result.dropna()
result
```

```python
for filename in os.listdir(input_folder):
    if filename.endswith(".csv"):
        if "UMAFall_Subject_13" in filename:
            continue
        file_path = os.path.join(input_folder, filename)
        df = pd.read_csv(file_path, header=None)
        df = df[~df.iloc[:, 0].astype(str).str.strip().str.startswith('%')]
        df.reset_index(drop=True, inplace=True)
        df = df[0].str.split(';', expand=True)
        df.drop(columns=[0], inplace=True)
        df.columns = ['Time Stamp', 'X-Axis', 'Y-Axis', 'Z-Axis', 'Sensor Type', 'Sensor ID']        

        df['Sensor ID'] = pd.to_numeric(df['Sensor ID'], errors='coerce')
        df = df[df['Sensor ID'].isin([1, 2])]
        df['Time Stamp'] = pd.to_numeric(df['Time Stamp'], errors='coerce')
        df['X-Axis'] = pd.to_numeric(df['X-Axis'], errors='coerce')
        df['Y-Axis'] = pd.to_numeric(df['Y-Axis'], errors='coerce')
        df['Z-Axis'] = pd.to_numeric(df['Z-Axis'], errors='coerce')
        df['Sensor Type'] = pd.to_numeric(df['Sensor Type'], errors='coerce')
        if df[df['Sensor ID'] == 1].empty or df[df['Sensor ID'] == 2].empty:
            print(f"Skipping {filename} as it does not contain both Sensor ID 1 and 2")
            continue

        df0 = df[df['Sensor Type'] == 0]
        df1 = df[df['Sensor Type'] == 1]
        df2 = df[df['Sensor Type'] == 2]
        df0.reset_index(drop=True, inplace=True)
        df1.reset_index(drop=True, inplace=True)
        df2.reset_index(drop=True, inplace=True)

        df0_1 = df0[df0['Sensor ID'] == 1]
        df0_2 = df0[df0['Sensor ID'] == 2]
        df1_1 = df1[df1['Sensor ID'] == 1]
        df1_2 = df1[df1['Sensor ID'] == 2]
        df2_1 = df2[df2['Sensor ID'] == 1]
        df2_2 = df2[df2['Sensor ID'] == 2]

        df0_1.reset_index(drop=True, inplace=True)
        df0_2.reset_index(drop=True, inplace=True)
        df1_1.reset_index(drop=True, inplace=True)
        df1_2.reset_index(drop=True, inplace=True)
        df2_1.reset_index(drop=True, inplace=True)
        df2_2.reset_index(drop=True, inplace=True)

        df0_1.columns = ['Time Stamp', 'A1 X-Axis', 'A1 Y-Axis', 'A1 Z-Axis', 'Sensor Type', 'Sensor ID']
        df0_2.columns = ['Time Stamp', 'A2 X-Axis', 'A2 Y-Axis', 'A2 Z-Axis', 'Sensor Type', 'Sensor ID']
        df1_1.columns = ['Time Stamp', 'G1 X-Axis', 'G1 Y-Axis', 'G1 Z-Axis', 'Sensor Type', 'Sensor ID']
        df1_2.columns = ['Time Stamp', 'G2 X-Axis', 'G2 Y-Axis', 'G2 Z-Axis', 'Sensor Type', 'Sensor ID']
        df2_1.columns = ['Time Stamp', 'M1 X-Axis', 'M1 Y-Axis', 'M1 Z-Axis', 'Sensor Type', 'Sensor ID']
        df2_2.columns = ['Time Stamp', 'M2 X-Axis', 'M2 Y-Axis', 'M2 Z-Axis', 'Sensor Type', 'Sensor ID']
        
        df0_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
        df0_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
        df1_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
        df1_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
        df2_1.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)
        df2_2.drop(columns=['Sensor Type', 'Sensor ID', 'Time Stamp'], inplace=True)

        result = pd.concat([df0_1, df1_1, df2_1, df0_2, df1_2, df2_2], axis=1)
        result.insert(0, 'Time Stamp', range(1, 1 + len(result)))
        result = result.dropna()
        if any(substring in filename for substring in ["backwardFall", "forwardFall", "lateralFall"]):
            result['ActivityLabel'] = 1
        else:
            result['ActivityLabel'] = 0 
        
        new_file_path = os.path.join(output_folder, f"processed_{filename}")
        result.to_csv(new_file_path, index=False)
        print(f"Processed {filename} and saved as {new_file_path}")
```

```python
dfs = []
for filename in os.listdir(output_folder):
    if filename.endswith(".csv"):
        file_path = os.path.join(output_folder, filename)
        df = pd.read_csv(file_path)
        dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.drop(columns=['Time Stamp', 'M1 X-Axis', 'M1 Y-Axis', 'M1 Z-Axis', 'M2 X-Axis', 'M2 Y-Axis', 'M2 Z-Axis'], inplace=True)
merged_df.insert(0, 'Time Stamp', range(1, 1 + len(merged_df)))
merged_df.to_csv(combined_file, index=False)
print(f"All CSV files merged and saved as {combined_file}")
```
